### Hyperparameter Grid search

```python
# import the modules
from sklearn.model_selection import GridSearchCV, KFold
from tensorflow.keras.layers import Dense
from tensorflow.keras import Model
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from tensorflow.keras.optimizers import Adam
```

```python
# define the model
def MyModel():
    inputs = tf.keras.Input(shape=(8))
    x = Dense(8, kernel_initializer='normal', activation='relu')(inputs)
    x = Dense(4, kernel_initializer='normal', activation='relu')(x)
    outputs = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=inputs, outputs=outputs, name='Deep_Model')
    
    optimizer = Adam(lr = 0.01)
    loss = 'binary_crossentropy'
    model.compile(loss=loss, optimizer=optimizer, metrics=['accuracy'])
    return model
    
```

```python
model = MyModel()
model.summary()
```

```python
# set a seed
seed = 1
np.random.seed(seed)


# define the model
def MyModel(lr, n1, n2, activation, init):
    inputs = tf.keras.Input(shape=(8))
    x = Dense(n1, kernel_initializer=init, activation=activation)(inputs)
    x = Dense(n2, kernel_initializer=init, activation=activation)(x)
    outputs = Dense(1, activation='sigmoid')(x)
    model = Model(inputs=inputs, outputs=outputs, name='Deep_Model')
    
    optimizer = Adam(lr = lr)
    loss = 'binary_crossentropy'
    model.compile(loss=loss, optimizer=optimizer, metrics=['accuracy'])
    return model

# create KerasClassifier model
model = KerasClassifier(build_fn=MyModel, verbose=0)

# set the grid search parameters
batch_size = [8, 16, 32]
epochs = [10, 50, 100]
lr = [0.001, 0.005, 0.01]
n1 = [4, 8, 16, 32]
n2 = [2, 4, 8, 16]
activation = ['relu', 'softmax', 'tanh', 'linear']
init = ['uniform', 'normal', 'zero']

# make a dictionary of grid search parameters
params = dict(batch_size=batch_size, epochs=epochs, activation=activation, init=init, lr=lr, n1=n1, n2=n2)

# make and fit the grid search model
gridModel = GridSearchCV(estimator=model, param_grid=params, cv=KFold(random_state=seed), refit=True, verbose=10)
gridSearchResults = gridModel.fit(standardizeData, y)

# print some of the results
print('Best: {} Using: {}'.format(gridSearchResults.best_score_, gridSearchResults.best_params_))
```

```python
means = gridSearchResults.cv_results_['mean_test_score']
stds = gridSearchResults.cv_results_['std_test_score']
parameters = gridSearchResults.cv_results_['params']
results = list(zip(means, stds, parameters))
results.sort(reverse=True)
for i in range(0, 10):
    print(results[i])
```
